```python
import requests as rq
import pandas as pd
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
# Need user agent info to access secs API.
headers = {'User-Agent': 'elijahworks@gmail.com'}

# CIK is an identification number for an company.
CIK = "0001318605"

# URL to Teslas financial statements
url = f"https://data.sec.gov/api/xbrl/companyfacts/CIK{CIK}.json"

# Request api to access to data. 
response = rq.get(url, headers=headers)

# Convert data to json format.
company_data = response.json()

# Check if api gives acess to the data.
if response.status_code == 200: print("Access grated")
else: print("Access NOT granted")

# How to access the common equity data!
concept_data1 = company_data['facts']['dei']['EntityCommonStockSharesOutstanding']['units']['shares']

pprint(concept_data1)
```

```python
years = [Y for Y in range(2016, 2023)]

common_equity = []
shares_outstanding = []

concepts = ['EntityCommonStockSharesOutstanding', 'EntityPublicFloat']
stock_price = []    

for concept in concepts:
    # Get Data into a JSON dictionary.
    concept_data = company_data['facts']['dei'][concept]['units']
    if concept == 'EntityCommonStockSharesOutstanding':
        concept_data2 = company_data['facts']['dei'][concept]['units']['shares']
        pprint(concept_data2)
    else: 
        concept_data2 = company_data['facts']['dei'][concept]['units']['USD']
        # pprint(concept_data2)
    
    for data in concept_data2:
        for year in years:
            try:
                if concept == 'EntityCommonStockSharesOutstanding':
                    if data['form'] == '10-K' and data['fy'] == year:
                        shares_outstanding.append(data['val'])
                elif concept == 'EntityPublicFloat':
                    if data['form'] == '10-K' and data['fy'] == year:
                        common_equity.append(data['val'])
            except: pass
  
```

```python

# Debugging
print("Year: ", end="")
for year in years: print(f"{year}, ", end="")
print()
print("common equity: ", end="")
for equity in common_equity: print(f"{equity}, ", end="")
print()
print("Shares: ", end="")
for share in shares_outstanding: print(f"{share}, ", end="")
print()
```
